```python
def make(filename,tag):
    
    file='D:\\사용자\\USER\\Desktop\\데이터\\최종2\\'+filename+'.csv'
    
    df1 = pd.read_csv(file)
    df1 = pd.DataFrame({'구':df1['구'], tag:df1['평균'] })
    return df1



```

```python
df1=make('서울건강','건강')
df2=make('서울교육','교육')
df3=make('서울상권','상권')
df4=make('서울생활','생활')
df5=make('서울스포츠','스포츠')
df6=make('서울안전','안전')
df7=make('서울여가','여가')
df8=make('서울체육','체육')
df9=make('서울환경','환경')


   
df10 = pd.read_csv('D:\\사용자\\USER\\Desktop\\데이터\\최종2\\서울집값.csv')
```

```python
m1=pd.merge(df1, df2 ,on="구")
m2=pd.merge(df3, df4 ,on="구")
m3=pd.merge(df5, df6 ,on="구")
m4=pd.merge(df7, df8 ,on="구")

m5=pd.merge(m1, m2 ,on="구")
m6=pd.merge(m3, m4 ,on="구")

m7=pd.merge(m5, m6 ,on="구")
m8=pd.merge(m7, df9 ,on="구")
m9=pd.merge(m8, df10 ,on="구")
m9
```
