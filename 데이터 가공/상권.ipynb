```python
mac6 = shop[shop["상권업종소분류명"].notnull() &shop['상권업종소분류명'].str.contains('편의점')].copy()
mac6.loc[mac6['상권업종소분류명'].str.contains('편의점'), '브랜드명'] = '편의점'
df_mac6 = pd.DataFrame(
    mac6.groupby(['시군구명', '브랜드명'])['상권업종소분류명'].count()
).reset_index()
df_mac6.columns = ['구', '브랜드명', '편의점']
df_mac6 = df_mac6.sort_values(by='편의점',ascending=False)
df_mac6 = df_mac6.reset_index(drop=True)
df_mac6 = df_mac6.drop(columns=['브랜드명'], axis=1)


df=pd.read_csv(r'D:\사용자\USER\Desktop\데이터\건강\서울총인구.csv', encoding='cp949', thousands=',')
df['구']=''
name_split = df["행정구역"].str.split(" ")
df["구"] = name_split.str.get(1)

df = df.drop(columns=['행정구역'], axis=1)

df2=pd.merge(df_mac6, df ,on="구")

df2
```

```python
def normal(df):
    data=list(df['편의점'])
    std_data = (data - np.mean(data,axis = 0))/ np.std(data,axis = 0)
    normalization=(std_data-min(std_data))/(max(std_data)-min(std_data))
    df['편의점']=normalization
    return df
```
