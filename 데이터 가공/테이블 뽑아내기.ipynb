```python
def make(df,values):
    df2 = pd.DataFrame({'시도':df['시도'], values:df[values] })
    df3 = df2.sort_values(by=values,ascending=False)
    df3=df3.reset_index()
    df3 = df3.drop(columns=['index'], axis=1)
    html = df3.to_html()
    print(html)
```
